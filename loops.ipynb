```python
# #Natural nos from 1 to 10 

# i=1
# while i<=10:
#     print(i)
#     i=i+1
                 
# #nos in reverse order 10 to 1
# j=10
# while j>=1:
#     print(j)
#     j=j-1

# #Even nos from 1 to 20
# k=2
# while k<=20:
#     print(k)
#     k=k+2
# #Odd nos from 1 to 20
# l=1
# while l<=10:
#     print(l)
#     l=l+2 
# #odd nos in reverse order from 20 to 1
# m=19
# while m>=1:
#     print(m)
#     m=m-2
# #digital sum of nos from 1 to n
# n=int(input("Enter a number: "))

# s=0
# while n>0:
#     r=n%10
#     s=s+r
#     n=n//10

# #reverse of a number
# n=int(input("Enter a number: "))
# sum=0
# while n>0:
#     d=n%10
#     sum=sum*10+d
#     n=n//10
# print("Reverse number is:",sum)
 
# #armstrong number
# n=int(input("Enter a number: "))
# temp=n
# sum=0
# while n>0:
#     r=n%10
#     sum=sum+r**3
#     n=n//10
# if temp==sum:
#     print("Armstrong number")
# #Factorial of a number
# n=int(input("Enter a number: "))
# fact=1
# while n>0:

#     fact=fact*n
#     n=n-1
# print("Factorial is:",fact)
# #Fibonacci series up to n terms
# n=int(input("Enter number of terms: "))
# a,b=0,1
# count=0
# while count<n:
#     print(a)
#     c=a+b
#     a=b
#     b=c
#     count+=1
# print("Digital sum is:",s)


for i in "Mango":
    print(i)


#All four types of functions
#1. Function with no argument and no return value
def greet():
    print("Hello!")
greet()
#2. Function with arguments but no return value
def greet(name):
    print("Hello, " + name + "!")
greet("Alice")
#3. Function with no argument but with return value
def get_greeting():
    return "Hello!"
message = get_greeting()
print(message)
#4. Function with arguments and return value
def add(a, b):
    return a + b
sum_result = add(5, 7)
print("Sum is:", sum_result)

#write a program to print the factorial of a number using four types of functions
#1. Function with no argument and no return value
def factorial_no_arg_no_return():
    n = int(input("Enter a number: "))
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    print("Factorial is:", fact)
factorial_no_arg_no_return()
#2. Function with arguments but no return value
def factorial_with_arg_no_return(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    print("Factorial is:", fact)
factorial_with_arg_no_return(5)
#3. Function with no argument but with return value
def factorial_no_arg_with_return():
    n = int(input("Enter a number: "))
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact
fact_result = factorial_no_arg_with_return()
print("Factorial is:", fact_result)

#4. Function with arguments and return value
def factorial_with_arg_with_return(n):
    fact = 1
    for i in range(1, n + 1):
        fact *= i
    return fact
fact_result = factorial_with_arg_with_return(6)
print("Factorial is:", fact_result) 
#write a program to swap numbers using all four types of functions
#1. Function with no argument and no return value
def swap_no_arg_no_return():
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    a, b = b, a
    print("After swapping: a =", a, ", b =", b)
swap_no_arg_no_return()
#2. Function with arguments but no return value
def swap_with_arg_no_return(a, b):      
    a, b = b, a
    print("After swapping: a =", a, ", b =", b)
swap_with_arg_no_return(3, 7)
#3. Function with no argument but with return value
def swap_no_arg_with_return():
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))
    return b, a
a, b = swap_no_arg_with_return()
print("After swapping: a =", a, ", b =", b)
#4. Function with arguments and return value
def swap_with_arg_with_return(a, b):
    return b, a
a, b = swap_with_arg_with_return(10, 20)
print("After swapping: a =", a, ", b =", b)


```
